```python
import numpy as np
import matplotlib.pyplot as plt
```

### Funcion Sigmoide

image.png

```python
def sigmoid(a):
  return 1/(1+np.exp(-a))
```

```python
test = np.linspace(10,-10, 100)

plt.plot(test, sigmoid(test))
```

### Funcion Escalonada

image.png

```python
def escalon(a):
  return np.piecewise(a, [a<0.0, a>0.0], [0,1])
  #Este metodo nos permite hacer un condicional desde adentro 
```

```python
plt.plot(test, escalon(test))
```

### ReLu

image.png

```python
def ReLU(a):
  return np.piecewise(a, [a<0.0, a>0.0], [0,lambda a:a])

#Aqui hay que utilizar el lambda por la documentacion del piecewise (no estamos en un for)
"""
Le pasamos un array de números (x), los cuales serán evaluados en una condicion [x<0.0, x >= 0.0], de acuerdo a la condición, 
si es True o False, retornará una función o un escalar. En caso que queramos aplicar una transformación al dato por medio de una función, se usa lambda

En el caso de la función step, no necesitamos usar lambda, ya que esta solo toma los valores 0 o 1.

En el caso de la relu, si usamos lambda para eL caso de que el dato sea mayor o igual que 0, retornar el propio valor. no podemos usar directamente x,
 es decir [0,x], porque aunque x es un escalar, este mismo va cambiando por cada valor en x del array de números.

"""
```

```python
plt.plot(test, ReLU(test))
```
